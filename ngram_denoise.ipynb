```python
def get_ngram_dict(ner_lst, min_window=1, max_window=15, split_char=None):
    """
    創建n-gram字典，統計字符出現頻率
    
    prd_lst - 包含要處理的字符串的列表
    min_window - 最小窗口大小（預設為1）
    max_window - 最大窗口大小（預設為15）
    split_char - 分隔字符（預設為None）
    
    output: 包含字符頻率的字典
    """
    
    char_freq_dict = defaultdict(int)
    
    for ner in ner_lst:
        if split_char:
            ner = ner.split(split_char)
            
        if len(ner) > max_window:
            
            for win in range(min_window, max_window+1):
                for i in range(len(ner)-win+1):
                    char = ner[i:i+win]
                    if split_char:
                        char = split_char.join(char)
                    char_freq_dict[char] += 1
        else:
            
            for win in range(min_window, len(ner)+1):
                for i in range(len(ner)-win+1):
                    char = ner[i:i+win]
                    if split_char:
                        char = split_char.join(char)
                    char_freq_dict[char] += 1
                
    return char_freq_dict

def get_english_ngrams(ner_list, min_window=1, max_window=15):

    eng_prd_list = [ner.strip().lower() for ner in ner_list if is_english(ner)]
    
    char_freq_dict = get_ngram_dict(eng_prd_list, min_window=min_window, max_window=max_window, split_char=" ")
    
    filtered_dict = {k: v for k, v in char_freq_dict.items() if not (is_number(k) or len(k) == 0)}
    
    filtered_dict = dict(sorted(filtered_dict.items(), key=lambda x: x[1], reverse=True))
    
    return filtered_dict

# get_english_ngrams(ner_list)
```

```python
def process_eng_ner(ner, char_freq_dict, threshold=1.2, stopwords=["the", "of", "an"]):
    
    if pd.isna(ner):
        return ner
    
    if isinstance(ner, str):
        ner = ner.split(",")
    elif isinstance(ner, list):
        pass
    else:
        raise TypeError("NER必須是要List或是String")
    
    ner_lst = [n.strip() for n in ner]
    new_ner_lst = []

    for volcab in ner_lst:
        if is_english(volcab):
            words = volcab.split(" ")
            candidate_chars = []

            for win in range(len(words), 0, -1):
                for i in range(len(words) - win + 1):
                    char = " ".join(words[i:i+win]).lower()

                    if is_number(char) or len(char) < 2 or char in stopwords:
                        continue

                    candidate_chars.append(char)

            final_char, max_score = char_compare(candidate_chars, char_freq_dict)

            if max_score >= threshold:
                new_ner_lst.append(final_char)
        else:
            new_ner_lst.append(volcab)

    new_ner_lst = list(set(new_ner_lst))
    
    return ",".join(new_ner_lst)
```
